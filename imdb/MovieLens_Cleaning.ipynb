```python
import duckdb
import numpy as np
import pandas as pd
```

```python
con = duckdb.connect(database=':memory:')
```

```python
movie_lens = pd.read_csv('movies_metadata.csv')
movie_lens['budget'].dtype
```

```python

movie_lens['budget'] = pd.to_numeric(movie_lens['budget'], errors='coerce') 
movie_lens['revenue'] = pd.to_numeric(movie_lens['revenue'], errors='coerce') 
movie_lens['popularity'] = pd.to_numeric(movie_lens['popularity'], errors='coerce') 
movie_lens['budget'] = movie_lens['budget'].fillna(0)
movie_lens['revenue'] = movie_lens['revenue'].fillna(0)
movie_lens['popularity'] = movie_lens['popularity'].fillna(0)



# movie_lens['budget'] = movie_lens['budget'].apply(lambda x: x if isinstance(x, int) else 0) 
# movie_lens['popularity'] = movie_lens['popularity'].apply(lambda x: x if isinstance(x, int) else 0) 
# movie_lens['revenue'] = movie_lens['revenue'].apply(lambda x: x if isinstance(x, int) else 0) 
```

```python
con.execute('''
    CREATE TABLE movie_lens_dataset AS SELECT * FROM movie_lens
''')
```

```python
con.execute("SELECT * FROM movie_lens_dataset").fetchdf()
```

```python
columns_to_drop = ['adult', 'original_language', 'overview', 'poster_path', 'release_date', 'runtime', 'status', 'tagline', 'video', 'vote_count', 'budget', 'revenue']

# drop the columns from the table
for column in columns_to_drop:
    con.execute(f'ALTER TABLE movie_lens_dataset DROP COLUMN {column}') 
```

```python
con.execute("""
    DELETE FROM movie_lens_dataset
    WHERE id IN (
        SELECT id
        FROM (
            SELECT id, ROW_NUMBER() OVER(PARTITION BY id ORDER BY id) AS rn
            FROM movie_lens_dataset
        ) t
        WHERE rn > 1
    )
""")

# set the id column as the primary key to ensure uniqueness
# con.execute('ALTER TABLE movie_lens_dataset ADD CONSTRAINT pk_movie_lens_dataset PRIMARY KEY (id)') 
```

```python
con.execute('''SELECT COUNT(*) FROM movie_lens_dataset WHERE vote_average = 0''').fetchdf()
```

```python
con.execute('''
    UPDATE movie_lens_dataset m1 
    SET vote_average = (
      SELECT AVG(vote_average) as yearly_mean 
      FROM movie_lens_dataset m2 
      WHERE m1.release_year = m2.release_year AND vote_average > 0 
      GROUP BY m2.release_year
    )
    WHERE vote_average = 0;
''').fetchdf()

```

```python
con.execute('''
    UPDATE movie_lens_dataset m1 
    SET popularity = (
      SELECT (SUM(popularity) - MIN(popularity) - MAX(popularity)) / CAST(COUNT(*)-2 as FLOAT) as trimmed_mean 
      FROM movie_lens_dataset m2 
      WHERE popularity > 0
    )
    WHERE popularity = 0;
'''
).fetchdf()
```
